```python
combined_df = doc_mapping.merge(hcpcs_raw_data,how='inner',left_on='NPI_DOC',right_on='NPI_DOC',suffixes=['_doc_mapping,','hcpcs_raw_data'])                                                      
```

```python
final_df = combined_df.head()[["NPI_DOC",
"PFNAME",
"ORG_NAME",
"DOC_SPECIALTY",
"DOC_CITY",
"DOC_STATE",
"SRVS_ASC_PHY","SRVS_IP_PHY","SRVS_OFF","SRVS_OP_PHY","HCPCS Code"]]
```

```python
final_df_with_hcpcs=final_df.merge(hcpcs_mapping,how='inner',left_on='HCPCS Code',right_on='HCPCS Code',suffixes=['_final_df','_hcpcs_mapping'])

final_df_with_hcpcs.head()
```

```python
final_df_with_hcpcs.columns = final_df_with_hcpcs.columns.str.replace(' ', '_')
# final_df_with_hcpcs.query()
```

```python
request = {
    "HCPCS_Class":["Cardiology"],
    "HCPCS_Subclass":["Clinical Cardiology"]
}

ls = [(k,v) for k,v in request.items()]
queries=[]
for item in ls:
    query = f"{item[0]} in {item[1]}"
    queries.append(query)
```

```python
final_query=" & ".join(queries)
final_query
```

```python
final_df_with_hcpcs.query(final_query)
```

```python
final_df_with_hcpcs.groupby(['NPI_DOC','PFNAME','ORG_NAME','DOC_SPECIALTY','DOC_CITY','DOC_STATE']).sum()
```

```python
final_op=final_df_with_hcpcs.groupby(['NPI_DOC','PFNAME','ORG_NAME','DOC_SPECIALTY','DOC_CITY','DOC_STATE']).sum().reset_index()
final_op.dtypes
```

```python
aggregated_op=[]
for index,row in final_op.iterrows():
    result = {
        'Physician NPI':row['NPI_DOC'],
        'Physician':row['PFNAME'],
        'Org Name': row['ORG_NAME'],
        'Speciality':row['DOC_SPECIALTY'],
        'City':row['DOC_CITY'],
        'State':row['DOC_STATE'],
        'ASC':row['SRVS_ASC_PHY'],
        'IP':row['SRVS_IP_PHY'],
        'OP':row['SRVS_OP_PHY'],
        'OFFICE':row['SRVS_OFF'],
        'Total': sum([row['SRVS_ASC_PHY'],row['SRVS_IP_PHY'],row['SRVS_OP_PHY'],row['SRVS_OFF']])
    }
    aggregated_op.append(result)
                      
```

```python
response = json.dumps(aggregated_op,indent=2)
```
